# 01 — Explore Dataset

Quick visual and statistical exploration of the raw image dataset before embedding extraction.

```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

DATASET_ROOT = Path('../data/raw/plantvillage')
PROCESSED_DIR = Path('../data/processed')
```

## Dataset structure

```python
classes = sorted([d for d in DATASET_ROOT.iterdir() if d.is_dir()])
class_counts = {c.name: len(list(c.rglob('*.jpg')) + list(c.rglob('*.png'))) for c in classes}

print(f'Total classes: {len(classes)}')
print(f'Total images: {sum(class_counts.values())}')
print()
for name, count in sorted(class_counts.items(), key=lambda x: -x[1]):
    print(f'  {name:50s}  {count:5d}')
```

## Class distribution

```python
fig, ax = plt.subplots(figsize=(14, 5))
names = list(class_counts.keys())
counts = list(class_counts.values())
ax.bar(names, counts)
ax.set_xticklabels(names, rotation=90, fontsize=8)
ax.set_ylabel('Images')
ax.set_title('Images per class')
plt.tight_layout()
plt.show()
```

## Sample images per class

```python
N_SHOW = min(6, len(classes))
fig, axes = plt.subplots(1, N_SHOW, figsize=(3 * N_SHOW, 3))

for ax, cls in zip(axes, classes[:N_SHOW]):
    images = list(cls.rglob('*.jpg')) + list(cls.rglob('*.png'))
    if images:
        img = Image.open(images[0]).convert('RGB').resize((224, 224))
        ax.imshow(img)
    ax.set_title(cls.name[:20], fontsize=8)
    ax.axis('off')

plt.tight_layout()
plt.show()
```

## Pre-extracted embeddings overview

```python
emb_path = PROCESSED_DIR / 'plantvillage_embeddings.npy'

if emb_path.exists():
    emb = np.load(emb_path)
    print(f'Embeddings shape : {emb.shape}')
    print(f'dtype            : {emb.dtype}')
    print(f'mean norm        : {np.linalg.norm(emb, axis=1).mean():.4f}')
    print(f'value range      : [{emb.min():.4f}, {emb.max():.4f}]')
else:
    print('Embeddings not found — run scripts/extract_embeddings.py first.')
```
